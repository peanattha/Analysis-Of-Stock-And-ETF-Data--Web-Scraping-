Web Scraping ข้อมูลรายชื่อหุ้นในตลาด SET

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import numpy as np
```

```python
list_url = ['NUMBER','A','B','C','D','E','F','G','H','I','J','K','M','N','L','O','P','Q','R','S','T','U','V','W','X','Y','Z']

info = pd.DataFrame()
for i in list_url: 
    url = 'https://classic.set.or.th/set/commonslookup.do?language=th&country=TH&prefix='+i

    r = requests.get(url)
    soup = BeautifulSoup(r.text, features="html.parser")
    table = soup.find('table')
    stock_name = pd.read_html(str(table))
    stock_name = stock_name[0]
    info = pd.concat([info,stock_name])
    # time.sleep(5)
info = info.reset_index()
info = info.drop(columns=['index'])
info
```

```python
info2 = info.copy()
```

```python
info2 = info2.rename(columns={"ชื่อย่อหลักทรัพย์":"securities abbreviation","ชื่อเต็มบริษัท / หลักทรัพย์จดทะเบียน":"Company full name","ตลาด":"market"})
info2
```

```python
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('—-incognito')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
browser = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
browser.get('http://siamchart.com/stock-info/2S/')
soup = browser.page_source
print(soup)
```

```python
dividend = pd.DataFrame()
num=1
for index, row in info2.iterrows():
  try:
    name = row['securities abbreviation']
    browser.get('http://siamchart.com/stock-info/'+name+'/')

    soup = BeautifulSoup(browser.page_source)
    table = soup.find_all('table')
    a = table[1]
    tr = a.find_all('tr')
    data = pd.DataFrame()
    for i in range (2,len(tr)):
      td = tr[i].find_all('td')
      new_row = {'ชื่อ':name,'วันที่':td[0].text, 'ปันผล':td[1].text, 'ราคาปิดก่อนปันผล':td[2].text, '%':td[3].text}
      data = data.append(new_row, ignore_index=True)
    dividend = pd.concat([dividend,data])
    time.sleep(5)
    print(num)
    num+=1
  except:
    num+=1
    print(num)
    continue

dividend
```

```python
dividend = dividend.reset_index()
dividend = dividend.drop(columns=['index'])
dividend
```
